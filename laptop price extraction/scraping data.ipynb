```python
import pandas as pd
import requests
import bs4
import pickle
import numpy as np
import re
import codecs
```

```python
df=pd.read_csv('urls.csv')
```

```python
prices=[]
description=[]
condition=[]
for i in range(df.shape[0]):
    try:      
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
        src = requests.get(df['urls'].iloc[i], headers=headers)
        soup = bs4.BeautifulSoup(src.text, 'html.parser')
        
        var=soup.find('div', class_='hidden-xs displayArrange')
        try:
            prices.append(var.find('span').text.strip())
        except:
            prices.append(np.nan)
        var2=soup.find('div', class_='productdetailsdata')
        try:
            var3=var2.find('div', class_='expandArea')
            condition.append(var3.find('span').text.strip())
        except:
            condition.append(np.nan)
        try:
            description.append(var2.find('div', class_='descriptionText').text.strip())
        except:
            description.append(np.nan)
    except:
        pass

store=[]
for i in range(df.shape[0]):
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
        src = requests.get(df['urls'].iloc[i], headers=headers)
        soup = bs4.BeautifulSoup(src.text, 'html.parser') 
        
        var2=soup.find('div', class_='productdetailsdata')
        var4=var2.find('div', class_='expandArea2')
        store.append(var4.find_all('span'))
    except:
        store.append(np.nan)
        
brand=[]
screen_size=[]
ram=[]
model=[]
product_type=[]
for i in range(len(store)):
    try:
        brand.append(store[i][0].text.strip())
    except:
        brand.append(np.nan)
    try:
        product_type.append(store[i][2].text.strip())
    except:
        product_type.append(np.nan)
    try:
        screen_size.append(store[i][3].text.split(' ')[2])
    except:
        screen_size.append(np.nan)
    try:
        ram.append(store[i][4].text.strip())
    except:
        ram.append(np.nan)
    try:
        model.append(store[i][6].text.strip())
    except:
        model.append(np.nan)
```

```python
print(len(prices))
print(len(condition))
print(len(description))
print(len(brand))
print(len(screen_size))
print(len(ram))
print(len(model))
print(len(product_type))
```

```python
d={'brand':brand,'condition':condition,'product_type':product_type,'model':model,'ram':ram,'screen_size':screen_size,'prices':prices
  ,'description':description}
data=pd.DataFrame(d)
```
